```python
def get_two_year(row):
    sub = matches[((matches.home_team == row['home_team']) & (matches.away_team == row['away_team'])) | ((matches.home_team == row['away_team']) & (matches.away_team == row['home_team']))]
    sub2 = sub.sort_values(by = 'date')
    
    sub3 = sub2[sub2['date'] < row['date']]
    home_team = row['home_team']
    away_team = row['away_team']
    home = 0
    away = 0
    for index, row2 in sub3.iterrows():
        winning_team = row2['winner']
        if winning_team == 'home':
            if row2['home_team'] == home_team:
                home += 1
            else:
                away += 1
        if winning_team == 'away':
            if row2['away_team'] == away_team:
                away += 1
            else:
                home += 1
    if (home + away) != 0:
        return home * 1.0 / (home + away) * 1.0
    else:
        if home == 0:
            return 0
        else:
            return 1
```

```python
################################################## Most recent version

def get_matchup_wins(row):
    sub = matches[((matches.home_team == row['home_team']) & (matches.away_team == row['away_team'])) | ((matches.home_team == row['away_team']) & (matches.away_team == row['home_team']))]
    sub2 = sub.sort_values(by = 'date')
    
    sub3 = sub2[sub2['date'] < row['date']]
    home_team = row['home_team']
    away_team = row['away_team']
    home = 0
    away = 0
    for index, row2 in sub3.iterrows():
        winning_team = row2['winner']
        if winning_team == 'home':
            if row2['home_team'] == home_team:
                home += 1
            else:
                away += 1
        if winning_team == 'away':
            if row2['away_team'] == away_team:
                away += 1
            else:
                home += 1
    if (home + away) != 0:
        return home * 1.0 / (home + away) * 1.0
    else:
        return -1 ###Haven't played each other before

sub = matches[matches['home_team'].isin(['Real Madrid', 'Barcelona'])].sort_values(by = 'date')
sub2 = sub[sub['away_team'].isin(['Real Madrid', 'Barcelona'])].sort_values(by = 'date')
sub2['func2'] = sub2.apply(lambda row: get_matchup_wins(row), axis = 1)
sub2.sort_values('date')
```

```python
############################################################################Aggregate Average

def aggregate_home(row):
    row = row.fillna(-1)
    bets = []
    bets.append(row['B365H'])
    bets.append(row['BWH'])
    bets.append(row['IWH'])
    bets.append(row['LBH'])
    bets.append(row['PSH'])
    bets.append(row['WHH'])
    bets.append(row['SJH'])
    bets.append(row['VCH'])
    bets.append(row['GBH'])
    bets.append(row['BSH'])

    counter = 0
    sum_bets = 0
    for item in bets:
        if item != -1:
            counter += 1
            sum_bets += item
    
    if counter == 0:
        return -1
    
    return sum_bets / counter
    
def aggregate_away(row):    
    row = row.fillna(-1)
    bets = []
    bets.append(row['B365A'])
    bets.append(row['BWA'])
    bets.append(row['IWA'])
    bets.append(row['LBA'])
    bets.append(row['PSA'])
    bets.append(row['WHA'])
    bets.append(row['SJA'])
    bets.append(row['VCA'])
    bets.append(row['GBA'])
    bets.append(row['BSA'])

    counter = 0
    sum_bets = 0
    for item in bets:
        if item != -1:
            counter += 1
            sum_bets += item
    
    if counter == 0:
        return -1
    
    return sum_bets / counter




def aggregate_tie(row):    
    row = row.fillna(-1)
    bets = []
    bets.append(row['B365D'])
    bets.append(row['BWD'])
    bets.append(row['IWD'])
    bets.append(row['LBD'])
    bets.append(row['PSD'])
    bets.append(row['WHD'])
    bets.append(row['SJD'])
    bets.append(row['VCD'])
    bets.append(row['GBD'])
    bets.append(row['BSD'])

    counter = 0
    sum_bets = 0
    for item in bets:
        if item != -1:
            counter += 1
            sum_bets += item
            
    if counter == 0:
        return -1    
        
    return sum_bets / counter

def bet_avg_winner(row):
    winner = min(row['home_bet_avg'], row['away_bet_avg'], row['tie_bet_avg'])
    if row['home_bet_avg'] == winner:
        return 'home'
    elif row['away_bet_avg'] == winner:
        return 'away'
    else:
        return 'tie'
    
def correct(row):
    if row['winner'] == row['bet_avg_winner']:
        return 1
    else:
        return 0


    

sub2 = matches.sort_values(by = 'date')
sub2['home_bet_avg'] = sub2.apply(lambda row: aggregate_home(row), axis = 1)
sub2['away_bet_avg'] = sub2.apply(lambda row: aggregate_away(row), axis = 1)
sub2['tie_bet_avg'] = sub2.apply(lambda row: aggregate_tie(row), axis = 1)
sub2['bet_avg_winner'] = sub2.apply(lambda row: bet_avg_winner(row), axis = 1)
sub2.sort_values('date')

sub2 = sub2[sub2['home_bet_avg'] != -1]

sub2['correct'] = sub2.apply(lambda row: correct(row), axis = 1)

sub2[['winner', 'bet_avg_winner', 'correct', 'home_bet_avg', 'away_bet_avg', 'tie_bet_avg', 'B365H', 'B365A', 'B365D','BWH','BWD', 'BWA', 'IWH', 'IWD', 'IWA', 'LBH', 'LBD', 'LBA']]

sub2['correct'].value_counts()
```

```python
def minl(l):
    min_value = min(l)
    return l.index(min_value)

def get_counts(row):
    
    home = 0
    away = 0
    tie = 0
    
    bet_houses = []
    
    row = row.fillna(-1)
    B365 = [row['B365H'], row['B365A'], row['B365D']]
    BW = [row['BWH'], row['BWA'], row['BWD']]
    IW = [row['IWH'], row['IWA'], row['IWD']]
    LB = [row['LBH'], row['LBA'], row['LBD']]
    PS = [row['PSH'], row['PSA'], row['PSD']]
    WH = [row['WHH'], row['WHA'], row['WHD']]
    SJ = [row['SJH'], row['SJA'], row['SJD']]
    VC = [row['VCH'], row['VCA'], row['VCD']]
    GB = [row['GBH'], row['GBA'], row['GBD']]
    BS = [row['BSH'], row['BSA'], row['BSD']]
    
    bet_houses.append(B365)
    bet_houses.append(BW)
    bet_houses.append(IW)
    bet_houses.append(LB)
    bet_houses.append(PS)
    bet_houses.append(WH)
    bet_houses.append(SJ)
    bet_houses.append(VC)
    bet_houses.append(GB)
    bet_houses.append(BS)
    
    for house in bet_houses:
        winner_index = minl(house)
        if winner_index == 0:
            home += 1
        elif winner_index == 1:
            away += 1
        else:
            tie += 1
            
    max_value = max(home, away, tie)
    if tie == max_value:
        return 'tie'
    elif away == max_value:
        return 'away'
    else:
        return 'home'
    
    

    

def correct_count(row):
    if row['winner'] == row['bet_avg_winner_count']:
        return 1
    else:
        return 0
    

sub2 = matches.sort_values(by = 'date')

sub2['bet_avg_winner_count'] = sub2.apply(lambda row: get_counts(row), axis = 1)
sub2.sort_values('date')


sub2['correct_count'] = sub2.apply(lambda row: correct_count(row), axis = 1)

sub2[['winner', 'bet_avg_winner_count', 'correct_count', 'B365H', 'B365A', 'B365D']]

sub2['correct_count'].value_counts()
```

```python
def trailing5_home(row):
    team = row['home_team']
    team_df =  matches[((matches.home_team == team) | (matches.away_team == team))]
    team_df = team_df.sort_values(by = 'date')
    team_df = team_df[team_df['date'] < row['date']] 
    if len(team_df) < 5:
        return -1.0
    
    team_df = team_df.tail(5)
    
    wins = 0.0
    losses = 0.0
    
    for index, row2 in team_df.iterrows():
        winning_team = row2['winner']
        if winning_team == 'home':
            if row2['home_team'] == team:
                wins += 1
            else:
                losses += 1
        elif winning_team == 'away':
            if row2['away_team'] == team:
                wins += 1
            else:
                losses += 1
        else:
            wins = wins ###Chuck ties
            
    if wins + losses == 0:
        return -1.0
    else:
        return wins / (wins + losses)
    
    
def trailing5_away(row):
    team = row['away_team']
    team_df =  matches[((matches.home_team == team) | (matches.away_team == team))]
    team_df = team_df.sort_values(by = 'date')
    team_df = team_df[team_df['date'] < row['date']] 
    if len(team_df) < 5:
        return -1.0
    
    team_df = team_df.tail(5)
    
    wins = 0.0
    losses = 0.0
    
    for index, row2 in team_df.iterrows():
        winning_team = row2['winner']
        if winning_team == 'home':
            if row2['home_team'] == team:
                wins += 1
            else:
                losses += 1
        elif winning_team == 'away':
            if row2['away_team'] == team:
                wins += 1
            else:
                losses += 1
        else:
            wins = wins ###Chuck ties
            
    if wins + losses == 0:
        return -1.0
    else:
        return wins / (wins + losses) 



sam = matches[((matches.home_team == 'Sevilla') | (matches.away_team == 'Sevilla'))].sort_values(by = 'date')
sam['trailing5_home'] = sam.apply(lambda row: trailing5_home(row), axis = 1)
sam['trailing5_away'] = sam.apply(lambda row: trailing5_away(row), axis = 1)
sam[['home_team', 'home_team_goal', 'away_team', 'away_team_goal', 'trailing5_home', 'trailing5_away']].head(20)
```
